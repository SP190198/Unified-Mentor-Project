```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('COVID clinical trials.csv')
df.head()
```

```python
null_values_count = df.isnull().sum()
null_values_count
```

```python
null_values_percentage = df.isnull().mean() * 100
null_values_percentage.round(2).astype(str) + "%"
```

```python
# Plotting
plt.figure(figsize=(10, 6))
null_values_percentage.plot(kind='bar')
plt.title('Count of Missing Values by Column')
plt.xlabel('Columns')
plt.ylabel('Missing Values Count')
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
null_values_percentage.plot(kind='bar')
plt.title('% of Missing Values by Column')
plt.xlabel('Columns')
plt.ylabel('% of Missing Values')
plt.show()
```

```python
results_first_posted = df[df['Results First Posted'].notnull()] #This step has been perfromed just to check what values are there in 'Results First Posted' column 
results_first_posted
```

```python
convert_to_date = ['Start Date', 'Primary Completion Date', 'Completion Date', 'First Posted', 'Results First Posted', 'Last Update Posted']
df[convert_to_date] = df[convert_to_date].apply(pd.to_datetime, errors = 'coerce')
df[['Start Date', 'Primary Completion Date', 'Completion Date', 'First Posted', 'Results First Posted', 'Last Update Posted']].info()
df[['Start Date', 'Primary Completion Date', 'Completion Date', 'First Posted', 'Results First Posted', 'Last Update Posted']].describe()
```

```python
check_for_unique_values = ['Status', 'Study Results', 'Conditions', 'Interventions', 'Locations', 'Study Type', 'Age', 'Phases']
unique_values = df[check_for_unique_values].nunique()
unique_values.astype(str) + " / " + str(len(df))
```

```python
countries = [ str(df['Locations'].iloc[i]).split(',')[-1] for i in range(df.shape[0])]
df['Country'] = countries
df['Country'].unique().tolist()
```

```python
df['Country'].value_counts().head(10).plot(kind='bar', title = 'Top 10 Countries Contribution')
```

```python
status_completed = df[df['Status'] == 'Completed']
countries = status_completed['Country'].value_counts().head(10)
countries.plot(kind='bar', title='Top 10 Countries with Status Completed', xlabel='Countries', ylabel='Status = Completed')

for index, value in enumerate(countries):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Filter DataFrame to include only the top 10 contributor countries
top_10_countries = df['Country'].value_counts().head(10).index
filtered_df = df[df['Country'].isin(top_10_countries)]

# Create a pivot table to count occurrences of each 'Study Result' for the top 10 countries
country_study_result_counts = filtered_df.pivot_table(
    index='Country',
    columns='Study Results',
    aggfunc='size',
    fill_value=0
)

# Plotting
country_study_result_counts.plot(kind='bar', stacked=True, figsize=(12, 8), colormap="viridis")
plt.title('Top 10 Contributor Countries by Study Result')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Study Result')
plt.show()
```
